```python
# find and remove the missing values
data.dropna(axis=1, how='all', inplace=True)
data
```

### Handling the NAN values 

```python
#Replace NAN values
data.fillna({'search_type':'none'}, inplace=True)

# checking Nan is replaced with None
data 
```

```python
#Converting date and time from object to date & time
data['stop_date'] = pd.to_datetime(data['stop_date'],format = "%Y-%m-%d")
data['stop_time'] = pd.to_datetime(data['stop_time'],format = "%H:%M:%S")
```

## Connect to MySQL

```python
import pymysql
```

```python
#SQL Connection
connection = pymysql.connect(host = '127.0.0.1',user='root',passwd='Haripriya@23')
cur = connection.cursor()
```

```python
#Creating policeledger database 
cur.execute("create database policeledger")
```

```python
#Use database policelog
cur.execute("use policeledger")
connection.commit()
```

```python
#Creating table 

cur.execute("""create table Police_Post_Logs(
id INT AUTO_INCREMENT PRIMARY KEY,
stop_date DATE,
stop_time TIME,
country_name VARCHAR(50),
driver_gender VARCHAR(10),
driver_age_raw INT,
driver_age INT,
driver_race VARCHAR(50),
violation_raw VARCHAR(100),
violation VARCHAR(100),
search_conducted BOOLEAN,
search_type VARCHAR(100),
stop_outcome VARCHAR(100),
is_arrested BOOLEAN,
stop_duration VARCHAR(50),
drugs_related_stop BOOLEAN,
vehicle_number VARCHAR(50))
""")
```

```python
#insert data into SQL Table

for index, row in data.iterrows():
    query = """
    INSERT INTO Police_Post_Logs (
        stop_date, stop_time, country_name, driver_gender, driver_age_raw,
        driver_age, driver_race, violation_raw, violation,
        search_conducted, search_type, stop_outcome, is_arrested,
        stop_duration, drugs_related_stop, vehicle_number
    )
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """
    
    values = (
        row['stop_date'], row['stop_time'], 
        row['country_name'], row['driver_gender'],
        row['driver_age_raw'], row['driver_age'], 
        row['driver_race'], row['violation_raw'],
        row['violation'], row['search_conducted'], 
        row['search_type'], row['stop_outcome'],
        row['is_arrested'], row['stop_duration'], 
        row['drugs_related_stop'], row['vehicle_number']
    )
    cur.execute(query, values)
```

```python
# Commit the changes
connection.commit()
```

```python
# Close connection
cur.close()
connection.close()
```
